```python
import pandas as pd

file_path = 'Files/response.xlsx'
df = pd.read_excel(file_path)
```

```python
import matplotlib.pyplot as plt

# Extract relevant data
grouped_data = df.groupby(['1. Do you prefer to use Generative AI (GenAI) for writing purposes in either academic or non-academic?', '0. Are you registered in either ENGL 355 or ENGL 357?']).size().unstack()

# Reverse the order of columns
grouped_data = grouped_data[['Yes', 'No']]

# Plot stacked bar chart
plt.figure(figsize=(10, 6))
ax = grouped_data.plot(kind='bar', stacked=True, color=['skyblue', 'lightcoral'], edgecolor='black')

plt.title('Do you prefer to use Generative AI for writing purposes?')
plt.xlabel('Preference for GenAI')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.legend(title='Are you in ENGL 355/357?', labels=['Yes', 'No'])

# Add data labels
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy() 
    ax.annotate(f'{int(height)}', (x + width/2, y + height/2), ha='center', va='center')

plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Patch

# Create a DataFrame with the required data
stacked_data = df.groupby(['4. How would you rate the impact of GenAI on your writing productivity on a scale of 1 to 5?', '0. Are you registered in either ENGL 355 or ENGL 357?']).size().unstack()

# Define custom colors for each category
colors = {'Yes': 'skyblue', 'No': 'lightcoral'}

# Plot stacked bar graph without grid
plt.figure(figsize=(10, 6))
ax = stacked_data.plot(kind='bar', stacked=True, color=[colors[col] for col in stacked_data.columns], edgecolor='black')

plt.title('Impact of GenAI on Writing Productivity')
plt.xlabel('Rating (1 to 5)')
plt.ylabel('Count')
plt.xticks(rotation=0)

# Add data labels
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy() 
    ax.annotate(f'{int(height)}', (x + width/2, y + height/2), ha='center', va='center')

# Create custom legend with specified colors and order
legend_order = ['Yes', 'No']
legend_elements = [Patch(facecolor=colors[col], edgecolor='black', label=col) for col in legend_order]

# Add custom legend
plt.legend(handles=legend_elements, title='Are you in ENGL 355/357?')

# Turn off the grid
ax.yaxis.grid(False)
ax.xaxis.grid(False)

plt.show()
```
